```python
ax=sns.countplot(x='Size',data=df)
```

```python
ax=sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

**Note : From above graph you can see that most of the people buys M-Size**

**Group By**

GroupBy in pandas is used to group data based on one or more columns in a DataFrame

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=S_Qty)
```

Note:From above Graph you can see that most of the Quantity buys M-Size in the sales

**COURIER STATUS**

```python
sns.countplot(x='Courier Status',data=df,hue='Status')
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=df,x='Courier Status',hue='Status')
plt.show()
```

```python
#histogram
df['Size'].hist()
```

```python
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=20,edgecolor='Black')
plt.xticks(rotation=90)
#plt.yticks(rotation=180)
plt.show()
```

Note : From the above Graph you can see that most of the buyers are T-Shirt

```python
#Checking B2B data using piechary
B2B_check=df['B2B'].value_counts()

#plot the piechart
plt.pie(B2B_check,labels=B2B_check,autopct='%1.1f%%')
plt.show()
```

```python
#Checking B2B data using piechary
B2B_check=df['B2B'].value_counts()

#plot the piechart
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
plt.show()
```

Note : From the above chart we can see that maximum i.e, 99.3% buyers are retailers and 0.8% are B2B buyers

```python
#prepare data for scatter plot
x_data=df['Category']
y_data=df['Size']

plt.scatter(x_data,y_data)
plt.xlabel("Category")
plt.ylabel('Size')
plt.title('Scatter plot')
plt.show()
```

```python
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
#top 10 states
top_10_states=df['ship-state'].value_counts().head(10)
#plot the count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

Note: From the above graph you can see that most of the buyers are from Maharastra state

Conclusion:
**The data analysis reveals that the business has significant customer base in Maharastra state, mainly serves retailers, fulfills the orders through Amazon, experiences high demand T-Shirts, and sees M-size as the preferred choice among buyers.**
